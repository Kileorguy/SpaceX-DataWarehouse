```python
import pandas as pd
import os

from IPython.terminal.shortcuts.auto_match import raw_string_braces
```

```python
ROOT_PATH = "./staging_data/splitted/"
raw_csvs = os.listdir(ROOT_PATH)

dataframes = {}

for csv in raw_csvs:
    path = ROOT_PATH + csv
    key = csv.split(".")[0]
    dataframes[key] = pd.read_csv(path)
```

```python
for key, item in dataframes.items():
    print(f"-===== {key} =====-")
    print(item.duplicated().sum())
    print(item.isna().sum())
    print("------------------------")

    for col in item.columns:
        print(f"{col}: {item[col].nunique()}")

```
